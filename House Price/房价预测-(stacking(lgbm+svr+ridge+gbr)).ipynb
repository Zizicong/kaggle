## 1.业务理解

房价是一个人们非常关注的问题，影响它的因素很多，但是具体和那些有关呢，做完这个项目，就有个大概的了解了。本项目是kaggle上的机器学习竞赛，官方提供了爱荷华州艾姆斯市的房价数据集，用一定解释变量描述了住宅的各个方面特征，比赛要求参赛者预测每一套住宅的最终价格。但是在开始之前，需要先理解数据，大概了解各特征对应的含义。<p>
根据官方提供的数据说明资料，各项特征对应的含义：<p>
1.MSSubClass: Identifies the type of dwelling involved in the sale，就是住宅的类型，官方已经做过离散化处理了；<p>
2.MSZoning: Identifies the general zoning classification of the sale，大概理解为购房的目的（工业用房、农业还是其他居住类型）；<p>
3.LotFrontage: Linear feet of street connected to property；房子到街道的距离，可能离街道越近越贵？<p>
4.LotArea: Lot size in square feet；住房面积，当然是越大越贵啦！<p>
5.Street: Type of road access to property；街道类型（grave：碎石路，没有铺面；paved：有铺面的道路）<p>
6.Alley: Type of alley access to property；巷道，和街道类型类似，但是有空值类型<p>
7.LotShape: General shape of property；土地的形状，可以理解为地基的形状？给出有四种（Reg： Regular；IR：Slightly irregular；IR2：Moderately Irregular：R3	Irregular）；<p>
8.LandContour: Flatness of the property；地基的平坦度，四种类型：Lvl、Bnk、HLS、Low分别表示水平、有点陡、非常陡、极其陡，就是越来越不平坦；<p>
9.Utilities: Type of utilities available；公共设备，即水电气等，有四种类型AllPub、NoSewr、NoSeWa、ELO，分别表示齐全、水电气、只有水气、只有电；<p>
10.LotConfig: Lot configuration；土地的位置，有内部，角落等类型；<p>
11.LandSlope: Slope of property；土地坡度，和特征7有点类似，有轻微，中等，严重三种；<p>
12.Neighborhood: Physical locations within Ames city limits；在艾姆斯市的位置，大概可以理解为哪个小区；<p>
13.Condition1: Proximity to various conditions；具体的地段，比如靠近主干道、支线等等；<p>
14.Condition2: Proximity to various conditions (if more than one is present)；和13类似，但是是符合多项；<p>
15.BldgType: Type of dwelling；单身公寓、复式等；<p>
16.HouseStyle: Style of dwelling；单层，两层及其其他形式等等；<p>
17.OverallQual: Rates the overall material and finish of the house；房子的材料和装饰评估等级，从最好到最差10和等级，已经离散化；<p>
18.OverallCond: Rates the overall condition of the house；房子整体评估等级，从最好到最差10和等级，已经离散化；<p>
19.YearBuilt: Original construction date；房子建好的年份；<p>
20.YearRemodAdd: Remodel date (same as construction date if no remodeling or additions)；改型日期，即房子建好之后又进行改型的年份；<p>
21.RoofStyle: Type of roof；屋顶类型；<p>
22.Roof material；屋顶材料；<p>
23.Exterior1st: Exterior covering on house；建房子材料的类型，石棉瓦，砖砌，木房等等；<p>
24.MasVnrType: Masonry veneer type；表层砌体类型，和23类似；<p>
25.MasVnrArea: Masonry veneer area in square feet；贴面面积:砖石贴面面积，单位为平方英尺；<p>
26.ExterQual: Evaluates the quality of the material on the exterior；外部材料评估等级，好到差有五种，但是是类别变量；<p>
27.ExterCond: Evaluates the present condition of the material on the exterior；外部材料评估现状等级，即随着时间流逝积极其他因素，房子目前的材料现状，有好到差有五种，但是是类别变量；<p>
28.Foundation: Type of foundation；地基类型，有石头，木等等；<p>
29.BsmtQual: Evaluates the height of the basement；地窖高度；类别变量，按照大小分为5个等级；<p>
30.BsmtCond: Evaluates the general condition of the basement；地窖等级评估，六个等级；<p>
31.BsmtExposure: Refers to walkout or garden level walls；地窖采光度；<p>
32.BsmtFinType1: Rating of basement finished area；地下室完工面积等级；好到差、无等七个等级；<p>
33.BsmtFinSF1: Type 1 finished square feet；成品面积，类型1；<p>
34.BsmtFinType2: Rating of basement finished area (if multiple types)；和32类似，有多类型的；<p>
35.BsmtFinSF2: Type 2 finished square feet；成品面积，类型2；<p>
36.BsmtUnfSF: Unfinished square feet of basement area；未完成的地下室面积面积；<p>
37.TotalBsmtSF: Total square feet of basement area；地下室总面积；<p>
38.Heating: Type of heating；供热类型，比如暖气加热、炉子加热、壁炉等等；<p>
39.HeatingQC: Heating quality and condition；供热系统评估；<p>
40.CentralAir: Central air conditioning；有无中央空调；<p>
41.Electrical: Electrical system；电力系统类型；<p>
42.1stFlrSF: First Floor square feet；一楼面积；<p>
43.2ndFlrSF: Second floor square feet；二楼面积；<p>
44.LowQualFinSF: Low quality finished square feet (all floors)；低质量的楼房面积（没建好的）；<p>
45.GrLivArea: Above grade (ground) living area square feet；以上(地面)居住面积平方英尺；<p>
46.BsmtFullBath: Basement full bathrooms；地下室全浴室；<p>
47.BsmtHalfBath: Basement half bathrooms；地下室半浴室；<p>
48.FullBath: Full bathrooms above grade；地面上浴室；<p>
49.HalfBath: Half baths above grade；浴室类型，仅具便池、面盆设备的卫生间；<p>
50.Bedroom: Bedrooms above grade (does NOT include basement bedrooms)；地面以上的卧室数量，不包括地下室的；<p>
51.Kitchen: Kitchens above grade；地面上的厨房数量；<p>
52.KitchenQual: Kitchen quality；厨房等级；<p>
53.TotRmsAbvGrd: Total rooms above grade (does not include bathrooms)；总房间数量，不包括浴室；<p>
54.Functional: Home functionality (Assume typical unless deductions are warranted)；家庭功能(假设是典型的，除非有必要进行扣减)，以典型为标准，按条件好坏进行扣减，8种类型；<p>
55.Fireplaces: Number of fireplaces；壁炉数量；<p>
56.FireplaceQu: Fireplace quality；壁炉质量；<p>
57.GarageType: Garage location；停车场类型，地下、家里、独立停车场等等；<p>
58.GarageYrBlt: Year garage was built；停车场建立年份；<p>
59.GarageFinish: Interior finish of the garage；停车场是否建好，有完成、大概完成、未完成、无停车场四个等级；<p>
60.GarageCars: Size of garage in car capacity；停车场容量，能停几辆车；<p>
61.GarageArea: Size of garage in square feet；停车场面积；<p>
62.GarageQual: Garage quality；停车场质量等级评估；<p>
63.GarageCond: Garage condition；停车场条件等级评估；<p>
64.PavedDrive: Paved driveway；行车道；扑倒、部分铺道、乱石路；<p>
65.WoodDeckSF: Wood deck area in square feet；木板面积；<p>
66.OpenPorchSF: Open porch area in square feet；门廊面积；<p>
67.EnclosedPorch: Enclosed porch area in square feet；封闭门廊面积，平方英尺；<p>
68.3SsnPorch: Three season porch area in square feet；三季门廊面积，平方英尺；<p>
69.ScreenPorch: Screen porch area in square feet；屏风门廊面积为平方英尺；<p>
70.PoolArea: Pool area in square feet；游泳池面积，平方英尺；<p>
71.PoolQC: Pool quality；游泳池质量等级；<p>
72.Fence: Fence quality；栅栏质量等级；<p>
73.MiscFeature: Miscellaneous feature not covered in other categories；杂项特性:其他类别没有涵盖的杂项特性，比如电梯，二楼停车场等；<p>
74.MiscVal: Value of miscellaneous feature；杂项功能的$价值；<p>
75.MoSold: Month Sold (MM)；出售的月份；<p>
76.YrSold: Year Sold (YYYY)；出售年份；<p>
77.SaleType: Type of sale；销售类型：如现金、贷款、15%首付等等；这一项应该比较关键；<p>
78.SaleCondition: Condition of sale；销售条件：正常的正常销售、非正常买卖、止赎、卖空。毗连土地购买、分配-两个相连的房产，各自的契约，通常是带车库单元的公寓、家庭成员之间的买卖、在最后一次评估时，部分房屋未完成(与新房屋相关)，具体的交易方式；<p>

根据以上提供的信息以及我们的常识可以知道，房子价格和房子大小(面积)、地段、设施、以及周边的路况都有关系；当然还包括买卖方式，贷款首付的方式肯定和全款不一样，不同年份也不一样等等，官方提供的信息非常的详细，需要我们仔细地去挖掘有价值的信息。

### 3.2.1 数值特征的缺失值处理

```python
missing_num = data_all[quantity].isnull().sum().sort_values(ascending=False)
missing_num = missing_num[missing_num>0]
missing_quant = (data_all[quantity].isnull().sum()/data_all[quantity].isnull().count()).sort_values(ascending=False)
missing_quant = missing_quant[missing_quant > 0] * 100
print("There are {} quantitative features".format(len(quantity)) + 
      " while {} quantitative features have  missing values.".format(missing_quant.shape[0]) )

missing_quant = pd.DataFrame({'Num' :missing_num, 'Percent' :missing_quant})
missing_quant
```

2.对于GarageYrBlt，编号为2576的没有GarageArea的信息，单独对其处理，对停车场建立年份赋值为0，其它缺失值按照LotFrontage一样，使用中位数进行填充；<p>
3.而对于其他的缺失，基本上都是面积类的，直接用0填充，认为其没有对应的面积。

```python
data_all.GarageArea[data_all.GarageArea.isnull()] #定位某一项缺失的数据，
```

```python
data_all['LotFrontage'] = data_all.groupby(['Neighborhood'])['LotFrontage'].apply(lambda x: x.fillna(x.median()))  # 1
data_all.at[2576, 'GarageYrBlt'] = 0
data_all['GarageYrBlt'] = data_all.groupby(['Neighborhood'])['GarageYrBlt'].apply(lambda x: x.fillna(x.median()))   # 2
for col in ['MasVnrArea','BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','BsmtFullBath','BsmtHalfBath','GarageCars',
            'GarageArea']:
    data_all[col] = data_all[col].fillna(0)                                                                           # 3
```

1.在业务理解的时候，我们发现，有一些评价类别变量我们是可以进行处理的（分等级），因此，我们将其按照评价好坏将其变成数值变量，数值越大表示评价越高；<p>
2.将出售年份、月份、MSSubClass转化成类别变量，根据官方文档解释来看，MSSubClass数字只是标记的意思，没有实际意义；因此我们将其转化为字符转（即类别信息），年月作类似处理；<p>

### 3.6 特征组合
选择合适的特征进行组合生成新的特征会有意想不到的效果，有一定的先验知识会有方向一点。<p>
1.将离散型的变量如GarageCars * GarageArea构造一个新的特征GarageScale表示车库规模，大概理解为能容纳的车子越多，车库面积越大，规模越大；<p>
2.将GarageQua + GarageCond构建一个新的特征表示车库的整体排名GarageOrdinal，车库质量越好，条件越好，整体排名越好。<p>
3.后面还有特征合并等，将同类型的特征进行合并，比如算出房子的廊道总面积、房子总面积等等；
特征组合的主要目的是希望能够发现各个特征之间的潜在联系，找出Magic Feature，能够很好的去描述模型，需要多次尝试，也是一个比较难的东西。

```python
# 根据对业务特征的理解，将原始特征组合生成新的特征
data_all['GarageScale'] = data_all['GarageCars'] * data_all['GarageArea'] # 1 能容纳的车子越多，车库面积越大，规模越大
data_all['GarageOrdinal'] = data_all['GarageQual'] + data_all['GarageCond'] # 2 车库质量越好，条件越好，整体排名越好
data_all['AllPorch'] = data_all['OpenPorchSF'] + data_all['EnclosedPorch'] + data_all['3SsnPorch'] + data_all['ScreenPorch'] #3 廊道总面积
data_all['ExterOrdinal'] = data_all['ExterQual'] + data_all['ExterCond'] #4 房子外部材料质量越好、 房子现在状况越好，房子外部整体情况就越好
data_all['KitchenCombined'] = data_all['KitchenQual'] * data_all['KitchenAbvGr'] #5 厨房质量越好、数量越多，厨房整体情况越好
data_all['FireplaceCombined'] = data_all['FireplaceQu'] * data_all['Fireplaces'] #6 壁炉质量越好、数量越多，壁炉整体情况越好
data_all['BsmtOrdinal'] = data_all['BsmtQual'] + data_all['BsmtCond'] #7 地下室质量越好、条件越好，地下室整体条件越好
data_all['BsmtFinishedAll'] = data_all['BsmtFinSF1'] + data_all['BsmtFinSF2'] # 地下室完成的总面积
data_all['AllFlrSF'] = data_all['1stFlrSF'] + data_all['2ndFlrSF'] #9 一楼面积和二楼面积总和
data_all['OverallCombined'] = data_all['OverallQual'] + data_all['OverallCond'] #10 房子的所有质量以及条件状况好坏
data_all['TotalFullBath'] = data_all['BsmtFullBath'] +  + data_all["FullBath"] #11 整体浴室面积
data_all['TotalHalfBath'] = data_all["HalfBath"] + data_all['BsmtHalfBath'] # 12 整体半浴室面积
data_all['TotalSF'] = data_all['AllFlrSF'] + data_all['TotalBsmtSF'] #13 房子总面积，所有层面积 + 地下室面积
data_all['YrBltAndRemod'] = data_all["YearRemodAdd"] + data_all['YearBuilt'] #14 房子建完的年份 + 改型的年份，建的越晚，改型越晚，房子就越新
```

### 3.7.2 数值特征筛选

选择相关性大于0.2的数值特征，这里的相关系数选择的是斯皮尔曼相关系数:<p>
1.关于spearman系数：斯皮尔曼相关性系数，通常也叫斯皮尔曼秩相关系数。“秩”，可以理解成就是一种顺序或者排序，那么它就是根据原始数据的排序位置进行求解，这种表征形式就没有了求皮尔森相关性系数时那些限制。而且，即便在变量值没有变化的情况下，也不会出现像皮尔森系数那样分母为0而无法计算的情况。另外，即使出现异常值，由于异常值的秩次通常不会有明显的变化（比如过大或者过小，那要么排第一，要么排最后），所以对斯皮尔曼相关性系数的影响也非常小，由于斯皮尔曼相关性系数没有那些数据条件要求，适用的范围就广多了。

```python
numeric = [f for f in data_all.columns if data_all.dtypes[f] != 'object']
numeric.remove('Id')
#计算斯皮尔曼相关系数
features = numeric
spr = pd.DataFrame()
spr['feature'] = features
spr['spearman'] = [data_all[f].corr(data_all['SalePrice'], 'spearman') for f in features]
spr = spr.sort_values('spearman')
```

#### 3.8.2 其它特征的变换

```python
# 集合数值特征
numerical = numeric_cols.values.tolist()
numerical.remove('SalePrice')
data_num = data_all[numerical].astype(float)
# 计算偏度
skewness = data_num.apply(lambda x: skew(x)).sort_values(ascending=False)
# 将偏度大于0.5的筛选出来，并作box-cox1p变换
skewness = skewness[abs(skewness) > 0.5]
from scipy.special import boxcox1p
skewed_features = skewness.index
lam = 0.15
for feat in skewed_features:
    data_num[feat] = boxcox1p(data_num[feat], lam)
```

Y = log(1+X)这个操作的真名应该是boxcox变换,用来降低X的skewness值,达到接近正态分布的目的. 

boxcox1p变换类似于yeojohnson变换，但是没有拟合lmbda值，可是认为给了一个经验值，我们这里给的值是0.15。

## 4 建模预测

```python
#载入模型
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor, BaggingRegressor
from sklearn.kernel_ridge import KernelRidge
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.linear_model import ElasticNet, ElasticNetCV
from sklearn.svm import SVR
from mlxtend.regressor import StackingCVRegressor
import lightgbm as lgb
from lightgbm import LGBMRegressor
from xgboost import XGBRegressor
```

```python
# 其他需要用到的包
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold, cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import scale
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
```

```python
# 定义均方误差损失函数
def rmse_cv(model):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring = "neg_mean_squared_error", cv = 5))
    return(rmse)
```

```python
from sklearn.model_selection import KFold
# 设定k折交叉验证
kf = KFold(n_splits=12, random_state=42, shuffle=True)
```

### 4.1 模型训练

```python
# Light Gradient Boosting Regressor
lightgbm = LGBMRegressor(objective='regression', 
                       num_leaves=6,         #叶子节点
                       learning_rate=0.01, 
                       n_estimators=7000,
                       max_bin=200,           #相比于其他学习器，LGBM有自己独特的分桶参数
                       bagging_fraction=0.8,  #随机采样比例
                       bagging_freq=4,        # 意味着每 k 次迭代执行bagging
                       bagging_seed=8,        
                       feature_fraction=0.2,  #随机采取特征比例
                       feature_fraction_seed=8,
                       min_sum_hessian_in_leaf = 11, #子节点所需的样本权重和(hessian)的最小阈值
                       verbose=-1,
                       random_state=42)

# Ridge Regressor
ridge_alphas = [1e-15, 1e-10, 1e-8, 9e-4, 7e-4, 5e-4, 3e-4, 1e-4, 1e-3, 5e-2, 1e-2, 0.1, 0.3, 1, 3, 5, 10, 15, 18, 20, 30, 50, 75, 100]
ridge = make_pipeline(RobustScaler(), RidgeCV(alphas=ridge_alphas, cv=kf))

# Support Vector Regressor
svr = make_pipeline(RobustScaler(), SVR(C= 20, epsilon= 0.008, gamma=0.0003))

# Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators=1000, learning_rate=.03, max_depth=3, max_features=.04, min_samples_split=4,
                                 min_samples_leaf=3, loss='huber', subsample=1.0, random_state=0) 


# 栈结合策略，使用ridge作为次级学习器
stack_gen = StackingCVRegressor(regressors=(lightgbm, svr, ridge, gbr),
                                meta_regressor=ridge,
                                use_features_in_secondary=True)
```

Light Gradient Boosting Regressor和Gradient Boosting Regressor:调参方法类似，LGBM相比于传统的梯度提升树,有着更快的运行速度,并有着更强的泛化性能.<p>
Ridge Regressor和Support Vector:二者都有权重参数,使用RobustScaler一定程度上减少outliers的影响.<p>

Stack:使用集成模型中的栈模型对初级学习器的结果进行再次学习，上述的前四个模型(lightgbm, svr, ridge, gbr)是初级学习器，这里选用的次级学习器是ridge，次级学习器用于训练的数据叫做次级训练集。次级训练集是在训练集上用初级学习器得到的。能够一定程度上提升模型泛化能力。

### 4.3 模型融合

```python
# 融合模型的目的是让最终模型有更好的泛化能力，因为最后的模型肯定会存在过拟合现象
def blended_predictions(X):
    return ((0.2 * ridge_model_full_data.predict(X)) + \
            (0.2 * svr_model_full_data.predict(X)) + \
            (0.2 * gbr_model_full_data.predict(X)) + \
            (0.2 * lgb_model_full_data.predict(X)) + \
            (0.2 * stack_gen_model.predict(np.array(X))))
```

```python
from sklearn.metrics import mean_squared_error
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

# 模型融合
blended_score = rmsle(y, blended_predictions(X))
scores['blended'] = (blended_score, 0)
print('RMSLE score on train data:')
print(blended_score)
```
